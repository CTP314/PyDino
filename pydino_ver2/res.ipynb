```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import MultipleLocator
```

```python
day_cnn = [178, 44,	228, 503, 81, 52, 234, 116, 193, 221, 108, 578,	147, 403, 570, 499, 232, 422, 254, 609]
x = range(1, 21)
c=np.mean(day_cnn)
x_major_locator = MultipleLocator(1)
ax = plt.gca()
ax.xaxis.set_major_locator(x_major_locator) 
plt.scatter(x,day_cnn, color="red", label="CNN")
plt.xlabel("Testing times", fontsize=15)
plt.ylabel("Score", fontsize=15)
plt.title("Day Result", fontsize=15)
plt.axhline(y=c,color="blue", label="CNN's average")
plt.legend()
plt.show()
```

```python
night_baseline = [719, 123, 111, 720, 126, 62, 44, 265, 78, 58, 53, 214, 725, 290, 59, 114, 83, 61, 747, 44]
night_cnn = [231, 82, 463, 136, 484, 135, 535, 67, 145, 150, 373, 645, 243, 121, 121, 173, 232, 460, 180, 662]
x = range(1, 21)
x_major_locator = MultipleLocator(1)
ax = plt.gca()
ax.xaxis.set_major_locator(x_major_locator) 
c=np.mean(night_cnn)
d=np.mean(night_baseline)
plt.scatter(x,night_cnn, color="red", label="CNN")
plt.scatter(x,night_baseline, color="orange", label="Baseline")
plt.xlabel("Testing times", fontsize=15)
plt.ylabel("Score", fontsize=15)
plt.title("Night Result", fontsize=15)
plt.axhline(y=c,color="blue",label="CNN's average")
plt.axhline(y=d,color="purple",label="Baseline's average")
plt.legend()
plt.show()
```
